```python
import csv , sqlite3

conn = sqlite3.connect("socioeconomic.db")
cur = conn.cursor()
```

```python
%sql sqlite:///socioeconomic.db
```

```python
import pandas as pd
df = pd.read_csv("Census_Data_-_Selected_socioeconomic_indicators_in_Chicago__2008___2012_20240711.csv")
df.columns = ["community_area_number","community_area_name","perecent_of_housing_crowded","percent_households_below_poverty","percent_aged_16+_unemployed","percent_aged_25+_without_high_school_diploma","percent_aged_under_18_or_over_64","per_capita_income","hardship_index"]
df.to_sql("chicago_socioeconomic_data",conn,if_exists='replace', index = False ,method = "multi" )
```

```python
%sql SELECT * FROM chicago_socioeconomic_data ;
```

```python
%sql select community_area_name from chicago_socioeconomic_data where per_capita_income > 60000;
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

income_vs_hardship = %sql SELECT per_capita_income, hardship_index FROM chicago_socioeconomic_data;
plot = sns.jointplot(x='per_capita_income',y='hardship_index', data=income_vs_hardship.DataFrame())
```

You can see that as Per Capita Income rises as the Hardship Index decreases. We see that the points on the scatter plot are somewhat closer to a straight line in the negative direction, so we have a negative correlation between the two variables.
